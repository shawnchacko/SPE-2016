```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
from pandas.plotting import andrews_curves
```

```python
data_df = pd.read_csv("/Users/shawnchacko/Desktop/SPE WA/SPE 2016 - Student Application Report.csv")
data_df
```

```python
data_df.rename(columns={'No. of Applicants': 'Curtin_Applications', 'Unnamed: 3': 'UWA_Applications', 'Total': 'Total_Applications', 'No. of Attendants':'No_of_attendants'}, inplace=True)
data_df
```

```python
data_final = data_df.iloc[1:]
data_final
```

```python
data_final.loc[(data_final['Topic'].isnull()), 'Topic']='No Event'
data_final[['Curtin_Applications', 'UWA_Applications']] = data_final[['Curtin_Applications','UWA_Applications']].fillna(value=0)
data_final
```

```python
data_final.loc[data_final['No_of_attendants'].isnull(),'No_of_attendants'] = data_final['No. of Invitations Sent']
data_final
```

```python
final_df = data_final.set_index(['Month'])
temp_df = final_df
final_df.plot(figsize=(15,10), x_compat=False)
plt.show()
```
